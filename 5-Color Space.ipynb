```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('avengers.jpg')
```

```python
plt.imshow(img)
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,15))

# OpenCv BGR
plt.subplot(1,2,1)
plt.imshow(img)
plt.xticks([]), plt.yticks([])

# Matplotlib RGB
plt.subplot(1,2,2)
plt.imshow(rgb_img)
plt.xticks([]),plt.yticks([])

plt.show()
```

```python
red = rgb_img.copy()
red[:,:,1]=0
red[:,:,2]=0

green = rgb_img.copy()
green[:,:,0]=0
green[:,:,2]=0

blue = rgb_img.copy()
blue[:,:,0]=0
blue[:,:,1]=0
```

```python
plt.figure(figsize=(15,20))

# red
plt.subplot(1,3,1)
plt.imshow(red)
plt.xticks([]),plt.yticks([])

# green
plt.subplot(1,3,2)
plt.imshow(green)
plt.xticks([]), plt.yticks([])

# blue
plt.subplot(1,3,3)
plt.imshow(blue)
plt.xticks([]), plt.yticks([])

plt.show()
```
