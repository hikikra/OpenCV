```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('gece.jpg')
```

```python
plt.imshow(img)
```

```python
# gürültü azaltalım
```

```python
img2 = cv2.bilateralFilter(img,9,100,100)
```

```python
plt.subplot(1,2,1), plt.imshow(img), plt.title('Orijinal')
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2), plt.imshow(img2), plt.title('Bilateral')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
#şimdi aydınlatalım
```

```python
parlak_img = cv2.convertScaleAbs(img2, beta=70)
```

```python
plt.figure(figsize=(15,20))
plt.subplot(121), plt.imshow(img2), plt.title('Orijinal')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(parlak_img), plt.title('Parlak')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
#konstrat-ını değiştirebiliriz
```

```python
kontrast_img = cv2.convertScaleAbs(img2, alpha=4)
```

```python
plt.figure(figsize=(15,20))

plt.subplot(121), plt.imshow(img2), plt.title('Orijinal')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(kontrast_img), plt.title('Kontrast')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
gray_img = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
parlak_gray = cv2.convertScaleAbs(gray_img, beta = 100)
plt.imshow(parlak_gray, cmap='gray')
```

```python
hist_gray = cv2.equalizeHist(gray_img)
```

```python
plt.figure(figsize=(15,20))

plt.subplot(121), plt.imshow(parlak_img,cmap='gray'), plt.title('Parlak Gri')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(hist_gray, cmap='gray'), plt.title('Histogram Eşitleme')
plt.xticks([]), plt.yticks([])

plt.show()
```
