```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('goz.jpg')
```

```python
plt.imshow(img)
```

```python
cekirdek = np.ones((8,8),np.float32)/36
img2 = cv2.filter2D(img,-1,cekirdek)
```

```python
plt.subplot(121), plt.imshow(img), plt.title('Orijinal')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(img2), plt.title('Ortalama Filtre')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
img3 = cv2.GaussianBlur(img,(7,7),0)
```

```python
plt.subplot(121), plt.imshow(img), plt.title('Orijinal')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(img3), plt.title('Gaussian Filtre')
plt.xticks([]), plt.yticks([])

plt.show()
```
