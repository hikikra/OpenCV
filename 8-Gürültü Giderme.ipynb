```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('gece.jpg')
```

```python
img2 = cv2.medianBlur(img,9)
```

```python
plt.subplot(1,2,1), plt.imshow(img), plt.title('Orijinal')
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2), plt.imshow(img2), plt.title('Median')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
img3 = cv2.imread('gece.jpg')
```

```python
img4 = cv2.bilateralFilter(img3,9,100,100)
```

```python
plt.subplot(1,2,1), plt.imshow(img3), plt.title('Orijinal')
plt.xticks([]), plt.yticks([])

plt.subplot(1,2,2), plt.imshow(img4), plt.title('Bilateral Filtered')
plt.xticks([]), plt.yticks([])

plt.show()
```
