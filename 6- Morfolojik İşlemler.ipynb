```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('A.jpg',0)
kernel = np.ones((5,5),np.uint8)
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
asindirma = cv2.erode(img,kernel,iterations = 1)
genisletme = cv2.dilate(img,kernel,iterations = 1)
```

```python
baslıklar=["image","aşındırma","genişletme"]
resimler=[img,asindirma,genisletme]
```

```python
plt.figure(figsize=(10,12))
for i in range(3):
  plt.subplot(1,3,i+1), plt.imshow(resimler[i])
  plt.title(baslıklar[i])
  plt.xticks([]),plt.yticks([])
plt.show()
```

```python
img2 = cv2.imread('biz.png')
kernel_10 = np.ones((2,2),np.uint8)
```

```python
asindirma_kodluyoruz = cv2.erode(img2,kernel_10,iterations = 1)
genisletme_kodluyoruz = cv2.dilate(img2,kernel_10,iterations = 1)
```

```python
images_2 = [img2, asindirma_kodluyoruz,genisletme_kodluyoruz]
```

```python
plt.figure(figsize=(8,11))
for i in range(3):
  plt.subplot(3,1,i+1), plt.imshow(images_2[i])
  plt.title(baslıklar[i])
  plt.xticks([]),plt.yticks([])
plt.show()
```
